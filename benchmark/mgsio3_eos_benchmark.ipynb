## 3. Reference Condition Validation

Compare calculated properties with Wolf et al. (2015) Table 3 values.

```python
# Reference values from Wolf et al. (2015) Table 3
# V₀ in Å³, K₀ in GPa, K'₀ dimensionless, γ₀, q, Θ₀ in K

# Physical constants
N_A = 6.02214076e23  # Avogadro number
Z = 4  # Formula units per unit cell

ref_values = {
    '0% Fe': {
        'V0_cell': 162.12e-30,  # m³ (cell volume)
        'K0': 262.3e9,          # Pa
        'K0_prime': 4.044,
        'gamma0': 1.675,
        'q': 1.39,
        'Theta0': 1000.0,       # K
        'M': 0.100387,          # kg/mol
    },
    '13% Fe': {
        'V0_cell': 163.16e-30,  # m³
        'K0': 243.8e9,          # Pa
        'K0_prime': 4.160,
        'gamma0': 1.400,
        'q': 0.56,
        'Theta0': 1000.0,       # K
        'M': (24.305*0.87 + 55.845*0.13 + 28.085 + 3*15.999) * 1e-3,  # kg/mol
    }
}

print("=" * 70)
print("REFERENCE CONDITION VALIDATION")
print("=" * 70)
print(f"\nComparing with Wolf et al. (2015) Table 3 at T = 300 K, P ~ 0")

for name, ref in ref_values.items():
    print(f"\n{'-'*50}")
    print(f"{name} End-member:")
    print(f"{'-'*50}")
    
    # Expected density from V₀
    V0_molar = ref['V0_cell'] * N_A / Z  # m³/mol
    rho_expected = ref['M'] / V0_molar
    
    # Get calculated values (at near-zero pressure)
    eos = eos_0Fe if name == '0% Fe' else eos_13Fe
    P_test = 1e5  # 1 bar
    T_test = 300.0
    
    rho_calc = eos.density(P_test, T_test)
    
    print(f"  Molar volume V₀:")
    print(f"    Reference:  {V0_molar*1e6:.4f} cm³/mol")
    print(f"    Calculated: {eos.params_0Fe['V0']*1e6 if name=='0% Fe' else eos.params_13Fe['V0']*1e6:.4f} cm³/mol")
    
    print(f"  Density at 300 K, ~0 GPa:")
    print(f"    Expected (from V₀): {rho_expected:.2f} kg/m³")
    print(f"    Calculated:         {rho_calc:.2f} kg/m³")
    print(f"    Difference:         {(rho_calc - rho_expected)/rho_expected * 100:.2f}%")
    
    print(f"\n  EoS Parameters:")
    print(f"    K₀:   {ref['K0']/1e9:.1f} GPa")
    print(f"    K'₀:  {ref['K0_prime']:.3f}")
    print(f"    γ₀:   {ref['gamma0']:.3f}")
    print(f"    q:    {ref['q']:.2f}")
    print(f"    Θ₀:   {ref['Theta0']:.0f} K")
```

## 4. Convergence Tests

Verify that the volume solver is accurate: P(V(P,T), T) ≈ P.

```python
# Test convergence across a range of P-T conditions
P_test = np.logspace(8, 11.2, 30)  # 0.1 MPa to 150 GPa
T_test = [300, 1000, 2000, 3000]

fig, axes = plt.subplots(1, 2, figsize=(14, 5), dpi=300)

for idx, eos in enumerate([eos_0Fe, eos_13Fe]):
    ax = axes[idx]
    title = f"{'Pure MgSiO₃' if idx == 0 else '13% Fe'}"
    
    for T in T_test:
        residuals = []
        P_valid = []
        
        for P in P_test:
            try:
                # Get volume at P, T
                V = eos._find_volume(P, T)
                # Calculate pressure at this volume
                P_calc = eos._total_pressure(V, T)
                # Compute relative residual
                residual = abs(P_calc - P) / P
                residuals.append(residual)
                P_valid.append(P)
            except RuntimeError:
                continue
        
        ax.semilogy(np.array(P_valid)/1e9, residuals, 'o-', label=f'$T = {T}$ K', markersize=4)
    
    ax.axhline(1e-4, color='r', ls='--', lw=1, alpha=0.7, label='0.01% target')
    ax.set_xlabel('Pressure [GPa]')
    ax.set_ylabel('$|P(V(P,T),T) - P| / P$')
    ax.set_title(f'{title}: Pressure Convergence')
    ax.legend(loc='upper right')
    ax.set_ylim(1e-16, 1e-2)
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nConvergence Summary:")
print(f"  Target tolerance: 0.01% (10⁻⁴)")
print(f"  All residuals should be well below the target line.")
```

## 5. Runtime Benchmarking

```python
# Benchmark runtime for different operations
P_bench = 100e9  # 100 GPa
T_bench = 2500   # 2500 K

n_calls = 100

print("=" * 70)
print("RUNTIME BENCHMARKING")
print("=" * 70)
print(f"\nConditions: P = {P_bench/1e9:.0f} GPa, T = {T_bench:.0f} K")
print(f"Number of calls per method: {n_calls}")

methods = ['density', 'specific_internal_energy', 'specific_entropy',
           'isobaric_heat_capacity', 'isochoric_heat_capacity',
           'thermal_expansion', 'adiabatic_gradient']

compositions = [('0% Fe', eos_0Fe), ('13% Fe', eos_13Fe), ('Mixed', eos_mixed)]

print(f"\n{'Method':<28} {'0% Fe [ms]':>12} {'13% Fe [ms]':>12} {'Mixed [ms]':>12}")
print("-" * 70)

for method in methods:
    times = []
    for name, eos in compositions:
        func = getattr(eos, method)
        
        start = time.perf_counter()
        for _ in range(n_calls):
            _ = func(P_bench, T_bench)
        elapsed = (time.perf_counter() - start) / n_calls * 1000  # ms per call
        times.append(elapsed)
    
    print(f"{method:<28} {times[0]:>12.3f} {times[1]:>12.3f} {times[2]:>12.3f}")
```

```python
# Grid benchmarking
print("\n" + "=" * 70)
print("GRID COMPUTATION BENCHMARK")
print("=" * 70)

# Create P-T grid
P_grid = np.linspace(20e9, 120e9, 20)  # 20-120 GPa
T_grid = np.linspace(300, 2500, 20)    # 300-2500 K
PP, TT = np.meshgrid(P_grid, T_grid)
n_total = PP.size

print(f"\nGrid size: {len(P_grid)} × {len(T_grid)} = {n_total} points")
print(f"P range: {P_grid[0]/1e9:.0f} - {P_grid[-1]/1e9:.0f} GPa")
print(f"T range: {T_grid[0]:.0f} - {T_grid[-1]:.0f} K")

for name, eos in compositions:
    start = time.perf_counter()
    for P, T in zip(PP.flat, TT.flat):
        _ = eos.density(P, T)
    elapsed = time.perf_counter() - start
    
    print(f"\n{name}:")
    print(f"  Total time: {elapsed:.3f} s")
    print(f"  Per point:  {elapsed/n_total*1000:.3f} ms")
    print(f"  Throughput: {n_total/elapsed:.0f} evaluations/s")
```

## 7. Thermodynamic Consistency Checks

Verify fundamental thermodynamic relations are satisfied.

```python
print("=" * 70)
print("THERMODYNAMIC CONSISTENCY CHECKS")
print("=" * 70)

# Check 1: C_P - C_V = α²TVK_T > 0
print("\n1. C_P - C_V > 0 check:")
print("-" * 50)

test_conditions = [
    (30e9, 1000, "Lower mantle entry"),
    (60e9, 2000, "Mid-lower mantle"),
    (100e9, 2500, "Deep mantle"),
]

for P, T, desc in test_conditions:
    for name, eos in [('0% Fe', eos_0Fe), ('13% Fe', eos_13Fe)]:
        Cp = eos.isobaric_heat_capacity(P, T)
        Cv = eos.isochoric_heat_capacity(P, T)
        diff = Cp - Cv
        
        status = "✓" if diff > 0 else "✗"
        print(f"  {desc}, {name}: C_P - C_V = {diff:.1f} J/(kg·K) {status}")
```

```python
# Check 2: Maxwell relation (∂S/∂P)_T = -(∂V/∂T)_P = -Vα
print("\n2. Maxwell relation check: (∂S/∂P)_T = -Vα")
print("-" * 50)
print("   Using numerical derivatives with ΔP = 0.1 GPa, comparing to -Vα")

dP = 0.1e9  # 0.1 GPa

for P, T, desc in test_conditions:
    for name, eos in [('0% Fe', eos_0Fe), ('13% Fe', eos_13Fe)]:
        # Numerical (∂S/∂P)_T
        S_plus = eos.specific_entropy(P + dP, T)
        S_minus = eos.specific_entropy(P - dP, T)
        dS_dP_num = (S_plus - S_minus) / (2 * dP)
        
        # Analytical -V α (specific, so -α/ρ)
        alpha = eos.thermal_expansion(P, T)
        rho = eos.density(P, T)
        dS_dP_analytical = -alpha / rho
        
        rel_err = abs(dS_dP_num - dS_dP_analytical) / abs(dS_dP_analytical) * 100 if dS_dP_analytical != 0 else 0
        status = "✓" if rel_err < 5 else "✗"
        
        print(f"  {desc}, {name}: numerical={dS_dP_num:.3e}, -α/ρ={dS_dP_analytical:.3e}, error={rel_err:.1f}% {status}")
```

```python
# Check 3: Positivity of α and K_T
print("\n3. Positivity check: α > 0 and K_T > 0")
print("-" * 50)

P_test = np.linspace(20e9, 120e9, 20)
T_test = [500, 1500, 2500]

all_positive = True
for name, eos in [('0% Fe', eos_0Fe), ('13% Fe', eos_13Fe)]:
    for T in T_test:
        for P in P_test:
            try:
                alpha = eos.thermal_expansion(P, T)
                # Get internal bulk modulus
                V = eos._find_volume(P, T)
                K_T = eos._isothermal_bulk_modulus(V, T)
                
                if alpha <= 0:
                    print(f"  WARNING: α ≤ 0 at P={P/1e9:.0f} GPa, T={T} K for {name}")
                    all_positive = False
                if K_T <= 0:
                    print(f"  WARNING: K_T ≤ 0 at P={P/1e9:.0f} GPa, T={T} K for {name}")
                    all_positive = False
            except:
                pass

if all_positive:
    print("  ✓ All α and K_T values positive across tested P-T range")
```

```python
# Check 4: Grüneisen parameter consistency
print("\n4. Grüneisen parameter: γ = αVK_T/C_V = αK_S/(ρC_P)")
print("-" * 50)

for P, T, desc in test_conditions:
    for name, eos in [('0% Fe', eos_0Fe), ('13% Fe', eos_13Fe)]:
        V = eos._find_volume(P, T)
        alpha = eos.thermal_expansion(P, T)
        K_T = eos._isothermal_bulk_modulus(V, T)
        Cv = eos._isochoric_heat_capacity(V, T)
        
        # From thermodynamic definition
        gamma_thermo = alpha * V * K_T / Cv
        
        # From EoS model (at current V)
        params = eos.params_0Fe if name == '0% Fe' else eos.params_13Fe
        gamma_model = eos._gruneisen_parameter(V, params)
        
        rel_diff = (gamma_thermo - gamma_model) / gamma_model * 100
        status = "✓" if abs(rel_diff) < 10 else "~"
        
        print(f"  {desc}, {name}: γ_thermo={gamma_thermo:.3f}, γ_model={gamma_model:.3f}, diff={rel_diff:+.1f}% {status}")
```

```python
# Volume crossover comparison
print("\n2. Density crossover between 0% Fe and 13% Fe:")
print("-" * 50)
print("   Wolf et al. note that Fe-bearing becomes denser than Fe-free above ~40 GPa at 300 K")

P_range = np.linspace(1e9, 120e9, 50)
T = 300.0

crossover_found = False
for i in range(len(P_range)-1):
    P1, P2 = P_range[i], P_range[i+1]
    
    rho_0Fe_1 = eos_0Fe.density(P1, T)
    rho_13Fe_1 = eos_13Fe.density(P1, T)
    rho_0Fe_2 = eos_0Fe.density(P2, T)
    rho_13Fe_2 = eos_13Fe.density(P2, T)
    
    # Check for crossover
    diff1 = rho_13Fe_1 - rho_0Fe_1
    diff2 = rho_13Fe_2 - rho_0Fe_2
    
    if diff1 * diff2 < 0:  # Sign change
        # Linear interpolation for crossover pressure
        P_cross = P1 + (P2 - P1) * (-diff1) / (diff2 - diff1)
        print(f"   Crossover found at P ≈ {P_cross/1e9:.1f} GPa")
        crossover_found = True
        break

if not crossover_found:
    print("   No crossover found in tested range (may be at very low or very high P)")

# Plot the density difference
fig, ax = plt.subplots(figsize=(10, 5), dpi=300)

rho_0Fe = np.array([eos_0Fe.density(P, T) for P in P_range])
rho_13Fe = np.array([eos_13Fe.density(P, T) for P in P_range])

ax.plot(P_range/1e9, rho_0Fe/1000, 'b-', label='0% Fe', lw=2)
ax.plot(P_range/1e9, rho_13Fe/1000, 'r-', label='13% Fe', lw=2)

ax.axvline(40, color='gray', ls='--', alpha=0.5, label='Expected crossover')

ax.set_xlabel('Pressure [GPa]')
ax.set_ylabel('Density [g/cm$^3$]')
ax.set_title('Density Crossover at 300 K')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Thermodynamic Consistency Tests <a name="consistency-sakai16"></a>

Verify that the EoS satisfies fundamental thermodynamic relations.

### Test: Cp ≥ Cv (thermodynamic stability)

```python
print("=" * 70)
print("THERMODYNAMIC CONSISTENCY CHECKS")
print("=" * 70)

# Check 1: C_P - C_V = α²TVK_T > 0
print("\n1. C_P - C_V > 0 check:")
print("-" * 50)

test_conditions = [
    (130e9, 2500, "Earth CMB"),
    (200e9, 3000, "Super-Earth mantle"),
    (300e9, 4000, "Deep super-Earth"),
    (500e9, 5000, "Extreme conditions"),
]

for P, T, desc in test_conditions:
    Cp = eos.isobaric_heat_capacity(P, T)
    Cv = eos.isochoric_heat_capacity(P, T)
    diff = Cp - Cv
    
    status = "✓" if diff > 0 else "✗"
    print(f"  {desc:<25s}: C_P - C_V = {diff:8.2f} J/(kg·K) {status}")
```

### Test: Maxwell relation (∂S/∂P)_T = -(∂V/∂T)_P = -Vα

```python
# Check 2: Maxwell relation (∂S/∂P)_T = -(∂V/∂T)_P = -Vα
print("\n2. Maxwell relation check: (∂S/∂P)_T = -Vα")
print("-" * 50)
print("   Using numerical derivatives with ΔP = 0.1 GPa, comparing to -Vα")

dP = 0.1e9  # 0.1 GPa

for P, T, desc in test_conditions:
    # Numerical (∂S/∂P)_T
    S_plus = eos.specific_entropy(P + dP, T)
    S_minus = eos.specific_entropy(P - dP, T)
    dS_dP_num = (S_plus - S_minus) / (2 * dP)
    
    # Analytical -V α (specific, so -α/ρ)
    alpha = eos.thermal_expansion(P, T)
    rho = eos.density(P, T)
    dS_dP_analytical = -alpha / rho
    
    rel_err = abs(dS_dP_num - dS_dP_analytical) / abs(dS_dP_analytical) * 100 if dS_dP_analytical != 0 else 0
    status = "✓" if rel_err < 5 else "✗"
    
    print(f"  {desc:<25s}: num={dS_dP_num:.3e}, -α/ρ={dS_dP_analytical:.3e}, err={rel_err:.1f}% {status}")
```

### Test: Entropy increases with temperature at constant pressure

```python
# Check 3: dS/dT > 0
print("\n3. Entropy increases with temperature: (∂S/∂T)_P > 0")
print("-" * 50)

P_test_s = [130, 300, 500, 1000]  # GPa
T_test_s = np.linspace(500, 5000, 50)

all_increasing = True
for P_gpa in P_test_s:
    S_vals = []
    for T in T_test_s:
        try:
            S_vals.append(eos.specific_entropy(P_gpa*GPa, T))
        except RuntimeError:
            S_vals.append(np.nan)
    S_arr = np.array(S_vals)
    dS = np.diff(S_arr[np.isfinite(S_arr)])
    if np.any(dS <= 0):
        all_increasing = False
        print(f"  P = {P_gpa} GPa: ✗ non-monotonic!")
    else:
        print(f"  P = {P_gpa} GPa: ✓ monotonically increasing")

if all_increasing:
    print("  ✓ All entropy tests passed")
```

### Test: Density increases with pressure at constant temperature

```python
# Check 4: dρ/dP > 0
print("\n4. Density increases with pressure: (∂ρ/∂P)_T > 0")
print("-" * 50)

T_vals_chk = [300, 2000, 4000]
P_range_chk = np.linspace(100*GPa, 1000*GPa, 100)

all_increasing = True
for T in T_vals_chk:
    rho_vals = []
    for P in P_range_chk:
        try:
            rho_vals.append(eos.density(P, T))
        except RuntimeError:
            rho_vals.append(np.nan)
    rho_arr = np.array(rho_vals)
    drho = np.diff(rho_arr[np.isfinite(rho_arr)])
    if np.any(drho <= 0):
        all_increasing = False
        print(f"  T = {T} K: ✗ non-monotonic!")
    else:
        print(f"  T = {T} K: ✓ monotonically increasing")

if all_increasing:
    print("  ✓ All density tests passed")
```

### Test: Positivity of α and K_T

```python
# Check 5: Positivity of α and K_T
print("\n5. Positivity check: α > 0 and K_T > 0")
print("-" * 50)

P_test = np.linspace(100*GPa, 800*GPa, 20)
T_test = [500, 2000, 4000]

all_positive = True
for T in T_test:
    for P in P_test:
        try:
            alpha = eos.thermal_expansion(P, T)
            V = eos._find_volume(P, T)
            K_T = eos._isothermal_bulk_modulus(V, T)
            
            if alpha <= 0:
                print(f"  WARNING: α ≤ 0 at P={P/GPa:.0f} GPa, T={T} K")
                all_positive = False
            if K_T <= 0:
                print(f"  WARNING: K_T ≤ 0 at P={P/GPa:.0f} GPa, T={T} K")
                all_positive = False
        except:
            pass

if all_positive:
    print("  ✓ All α and K_T values positive across tested P-T range")
```

### Test: Grüneisen parameter consistency γ = αVK_T/C_V

```python
# Check 6: Grüneisen parameter consistency
print("\n6. Grüneisen parameter: γ = αVK_T/C_V")
print("-" * 50)

print(f"\n  {'Condition':<25s} {'γ_model':>10s} {'γ_thermo':>10s} {'Diff (%)':>10s}")
print("  " + "-" * 60)

for P, T, desc in test_conditions:
    V = eos._find_volume(P, T)
    alpha = eos.thermal_expansion(P, T)
    K_T = eos._isothermal_bulk_modulus(V, T)
    Cv = eos._isochoric_heat_capacity_molar(V, T)
    
    # From EoS model
    gamma_model = eos._gruneisen_parameter(V)
    
    # From thermodynamic definition
    gamma_thermo = alpha * V * K_T / Cv
    
    rel_diff = (gamma_thermo - gamma_model) / gamma_model * 100
    status = "✓" if abs(rel_diff) < 5 else "~"
    
    print(f"  {desc:<25s} {gamma_model:>10.4f} {gamma_thermo:>10.4f} {rel_diff:>+9.2f}% {status}")
```

```python
# 5. Keane EoS derivative verification
print("\n5. Keane EoS derivative verification:")
print("-" * 50)

V0 = eos.params['V0']
K0 = eos._cold_bulk_modulus(V0)

# Numerical K₀' = dK/dP at V₀
dV = V0 * 1e-6
K_plus = eos._cold_bulk_modulus(V0 - dV)  # smaller V = higher P
K_minus = eos._cold_bulk_modulus(V0 + dV)
P_plus = eos._cold_pressure(V0 - dV)
P_minus = eos._cold_pressure(V0 + dV)

K0_prime_num = (K_plus - K_minus) / (P_plus - P_minus)

print(f"  K₀ = {K0/GPa:.2f} GPa (expect {eos.params['K0']/GPa:.1f})")
print(f"  K₀' (numerical) = {K0_prime_num:.4f} (expect {eos.params['K0_prime']:.3f})")
print(f"  Relative error:   {abs(K0_prime_num - eos.params['K0_prime'])/eos.params['K0_prime']*100:.4f}%")

# Keane's rule: K₀'/2 < K∞' < K₀' - 1
K0p = eos.params['K0_prime']
Kinf = eos.params['K_inf_prime']
print(f"\n  Keane's rule: K₀'/2 < K∞' < K₀' - 1")
print(f"    {K0p/2:.3f} < {Kinf:.3f} < {K0p - 1:.3f}")
keane_ok = (K0p / 2 < Kinf < K0p - 1)
print(f"    {'✓ Satisfied' if keane_ok else '✗ Violated'}")

# Thermodynamic constraints
print(f"\n  Thermodynamic constraints:")
print(f"    K∞' = {Kinf:.3f} > 5/3 = {5/3:.3f}: {'✓' if Kinf > 5/3 else '✗'}")
print(f"    γ∞  = {eos.params['gamma_inf']:.3f} > 2/3 = {2/3:.3f}: {'✓' if eos.params['gamma_inf'] > 2/3 else '✗'}")
```

## 3. Reference Condition Validation

Compare calculated properties at V₀ with Sokolova et al. (2022) Table 2 parameters.
Verify that the Kunc cold pressure and derived quantities match at the reference volume.

```python
print("=" * 70)
print("REFERENCE CONDITION VALIDATION")
print("=" * 70)

for name, eos in phases.items():
    p = eos.params
    V0 = p['V0']
    
    print(f"\n--- {name} ---")
    print(f"{'Parameter':<20s} {'Expected':>12s} {'Calculated':>12s} {'Match':>8s}")
    print("-" * 55)
    
    # Cold pressure at V₀ should be ~0
    P0 = eos._cold_pressure(V0)
    print(f"  {'P_cold(V₀) [GPa]':<20s} {'0.000':>12s} {P0/GPa:>12.4f} "
          f"{'✓' if abs(P0/GPa) < 0.001 else '✗':>8s}")
    
    # Cold bulk modulus at V₀ should be K₀
    K0_calc = eos._cold_bulk_modulus(V0)
    K0_exp = p['K0']
    diff_K = abs(K0_calc - K0_exp) / K0_exp * 100
    print(f"  {'K_cold(V₀) [GPa]':<20s} {K0_exp/GPa:>12.1f} {K0_calc/GPa:>12.2f} "
          f"{'✓' if diff_K < 0.1 else '✗':>8s}")
    
    # Cold energy at V₀ should be ~0
    E0 = eos._cold_energy(V0)
    print(f"  {'E_cold(V₀) [J/mol]':<20s} {'0.000':>12s} {E0:>12.4f} "
          f"{'✓' if abs(E0) < 0.01 else '✗':>8s}")
    
    # Grüneisen parameter at V₀
    gamma_calc = eos._gruneisen_parameter(V0)
    gamma_exp = p['gamma0']
    print(f"  {'γ(V₀)':<20s} {gamma_exp:>12.3f} {gamma_calc:>12.4f} "
          f"{'✓' if abs(gamma_calc - gamma_exp) < 0.001 else '✗':>8s}")
    
    # q parameter at V₀ (should be β for our Al'tshuler form)
    q_calc = eos._q_parameter(V0)
    q_exp = p['beta']
    print(f"  {'q(V₀)':<20s} {q_exp:>12.3f} {q_calc:>12.4f} "
          f"{'✓' if abs(q_calc - q_exp) < 0.001 else '✗':>8s}")
    
    # Einstein temperatures at V₀
    for i in [1, 2]:
        theta_calc = eos._einstein_temperature(V0, i)
        theta_exp = p[f'theta0{i}']
        print(f"  {'Θ'+str(i)+'(V₀) [K]':<20s} {theta_exp:>12.1f} {theta_calc:>12.2f} "
              f"{'✓' if abs(theta_calc - theta_exp) < 0.1 else '✗':>8s}")

print()
print("All reference checks verify that P_cold(V₀) = 0, K_cold(V₀) = K₀,")
print("E_cold(V₀) = 0, and model parameters are correctly reproduced.")
```

## 7. Thermodynamic Consistency Checks

Verify fundamental thermodynamic relations are satisfied.

### Test: Cp ≥ Cv (thermodynamic stability)

```python
print("=" * 70)
print("THERMODYNAMIC CONSISTENCY CHECKS")
print("=" * 70)

# Check 1: C_P - C_V = α²TVK_T > 0
print("\n1. C_P - C_V > 0 check:")
print("-" * 50)

test_conditions = [
    (1e5,    298.15, "Ambient"),
    (3*GPa,  1000,   "Shallow mantle"),
    (8*GPa,  1500,   "Mid-upper mantle"),
    (15*GPa, 2000,   "Transition zone"),
]

for name, eos in phases.items():
    print(f"\n  {name}:")
    for P, T, desc in test_conditions:
        try:
            Cp = eos.isobaric_heat_capacity(P, T)
            Cv = eos.isochoric_heat_capacity(P, T)
            diff = Cp - Cv
            status = "✓" if diff > 0 else "✗"
            print(f"    {desc:<25s}: C_P - C_V = {diff:8.2f} J/(kg·K) {status}")
        except RuntimeError:
            print(f"    {desc:<25s}: no convergence")
```

### Test: Maxwell relation (∂S/∂P)_T = -(∂V/∂T)_P = -Vα

```python
# Check 2: Maxwell relation (∂S/∂P)_T = -(∂V/∂T)_P = -Vα
print("\n2. Maxwell relation check: (∂S/∂P)_T = -α/ρ")
print("-" * 50)
print("   Using numerical derivatives with ΔP = 0.01 GPa")

dP = 0.01e9

for name, eos in phases.items():
    print(f"\n  {name}:")
    for P, T, desc in test_conditions:
        try:
            S_plus  = eos.specific_entropy(P + dP, T)
            S_minus = eos.specific_entropy(P - dP, T)
            dS_dP_num = (S_plus - S_minus) / (2 * dP)
            
            alpha = eos.thermal_expansion(P, T)
            rho = eos.density(P, T)
            dS_dP_analytical = -alpha / rho
            
            rel_err = abs(dS_dP_num - dS_dP_analytical) / abs(dS_dP_analytical) * 100
            status = "✓" if rel_err < 5 else "✗"
            
            print(f"    {desc:<25s}: num={dS_dP_num:.3e}, -α/ρ={dS_dP_analytical:.3e}, "
                  f"err={rel_err:.1f}% {status}")
        except RuntimeError:
            print(f"    {desc:<25s}: no convergence")
```

### Test: Entropy increases with temperature at constant pressure

```python
# Check 3: dS/dT > 0
print("\n3. Entropy increases with temperature: (∂S/∂T)_P > 0")
print("-" * 50)

P_test_s = [1, 5, 10, 15]  # GPa
T_test_s = np.linspace(400, 2000, 50)

all_increasing = True
for name, eos in phases.items():
    for P_gpa in P_test_s:
        S_vals = []
        for T in T_test_s:
            try:
                S_vals.append(eos.specific_entropy(P_gpa*GPa, T))
            except RuntimeError:
                S_vals.append(np.nan)
        S_arr = np.array(S_vals)
        dS = np.diff(S_arr[np.isfinite(S_arr)])
        if np.any(dS <= 0):
            all_increasing = False
            print(f"  {name}, P = {P_gpa} GPa: ✗ non-monotonic!")
        else:
            print(f"  {name}, P = {P_gpa} GPa: ✓ monotonically increasing")

if all_increasing:
    print("\n  ✓ All entropy tests passed")
```

### Test: Density increases with pressure at constant temperature

```python
# Check 4: dρ/dP > 0
print("\n4. Density increases with pressure: (∂ρ/∂P)_T > 0")
print("-" * 50)

T_vals_chk = [300, 1000, 2000]
P_range_chk = np.linspace(0.5*GPa, 20*GPa, 80)

all_increasing = True
for name, eos in phases.items():
    for T in T_vals_chk:
        rho_vals = []
        for P in P_range_chk:
            try:
                rho_vals.append(eos.density(P, T))
            except RuntimeError:
                rho_vals.append(np.nan)
        rho_arr = np.array(rho_vals)
        drho = np.diff(rho_arr[np.isfinite(rho_arr)])
        if np.any(drho <= 0):
            all_increasing = False
            print(f"  {name}, T = {T} K: ✗ non-monotonic!")
        else:
            print(f"  {name}, T = {T} K: ✓ monotonically increasing")

if all_increasing:
    print("\n  ✓ All density tests passed")
```

### Test: Positivity of α and K_T

```python
# Check 5: Positivity of α and K_T
print("\n5. Positivity check: α > 0 and K_T > 0")
print("-" * 50)

P_test = np.linspace(0.5*GPa, 15*GPa, 15)
T_test = [300, 1000, 2000]

all_positive = True
for name, eos in phases.items():
    for T in T_test:
        for P in P_test:
            try:
                alpha = eos.thermal_expansion(P, T)
                V = eos._find_volume(P, T)
                K_T = eos._isothermal_bulk_modulus(V, T)
                
                if alpha <= 0:
                    print(f"  WARNING: α ≤ 0 at {name}, P={P/GPa:.1f} GPa, T={T} K")
                    all_positive = False
                if K_T <= 0:
                    print(f"  WARNING: K_T ≤ 0 at {name}, P={P/GPa:.1f} GPa, T={T} K")
                    all_positive = False
            except:
                pass

if all_positive:
    print("  ✓ All α and K_T values positive across tested P-T range")
```

### Test: Grüneisen parameter consistency γ = αVK_T/C_V

```python
# Check 6: Grüneisen parameter consistency
print("\n6. Grüneisen parameter: γ = αVK_T/C_V")
print("-" * 50)

print(f"\n  {'Phase':<10s} {'Condition':<25s} {'γ_model':>10s} {'γ_thermo':>10s} {'Diff (%)':>10s}")
print("  " + "-" * 70)

for name, eos in phases.items():
    for P, T, desc in test_conditions:
        try:
            V = eos._find_volume(P, T)
            alpha = eos.thermal_expansion(P, T)
            K_T = eos._isothermal_bulk_modulus(V, T)
            Cv = eos._isochoric_heat_capacity(V, T)
            
            gamma_model = eos._gruneisen_parameter(V)
            gamma_thermo = alpha * V * K_T / Cv
            
            rel_diff = (gamma_thermo - gamma_model) / gamma_model * 100
            status = "✓" if abs(rel_diff) < 5 else "~"
            
            print(f"  {name:<10s} {desc:<25s} {gamma_model:>10.4f} {gamma_thermo:>10.4f} "
                  f"{rel_diff:>+9.2f}% {status}")
        except RuntimeError:
            print(f"  {name:<10s} {desc:<25s} {'—':>10s} {'—':>10s} {'—':>10s}")
```

## 4. Convergence Tests

```python
# Test convergence by checking P(V(P,T), T) ≈ P
# across the AIMD calibration range: 0-1200 GPa, 2200-14000 K
test_pressures = np.logspace(np.log10(1*GPa), np.log10(1200*GPa), 30)
test_temperatures = [2500, 3000, 5000, 8000, 10000, 14000]

fig, ax = plt.subplots(figsize=(10, 5), dpi=300)

for T in test_temperatures:
    errors = []
    valid_P = []
    for P in test_pressures:
        try:
            V = eos._find_volume(P, T)
            P_calc = eos._rtpress.eval_press(V, T) * GPa  # GPa → Pa
            residual = abs(P_calc - P) / P
            errors.append(residual)
            valid_P.append(P)
        except RuntimeError:
            continue
    ax.semilogy(np.array(valid_P)/GPa, errors, 'o-', label=f'$T = {T}$ K',
                markersize=4)

ax.axhline(1e-4, color='r', ls='--', lw=1, alpha=0.7, label='0.01% target')
ax.set_xlabel('Pressure [GPa]')
ax.set_ylabel('$|P(V(P,T),T) - P| / P$')
ax.set_title('Wolf18 Liquid MgSiO₃: Pressure Convergence')
ax.legend(loc='upper right', fontsize=9)
ax.set_ylim(1e-16, 1e-2)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nConvergence Summary:")
print(f"  Target tolerance: 0.01% (10⁻⁴)")
print(f"  All residuals should be well below the target line.")
```

## 7. Thermodynamic Consistency Checks

### Test: Cp ≥ Cv (thermodynamic stability)

```python
print("=" * 70)
print("THERMODYNAMIC CONSISTENCY CHECKS")
print("=" * 70)

# Check 1: C_P - C_V = α²TVK_T > 0
print("\n1. C_P - C_V > 0 check:")
print("-" * 50)

test_conditions = [
    (50e9,   3000,  "Deep Earth magma ocean"),
    (200e9,  5000,  "Super-Earth mantle"),
    (500e9,  8000,  "Deep super-Earth"),
    (1000e9, 12000, "Sub-Neptune interior"),
]

for P, T, desc in test_conditions:
    Cp = eos.isobaric_heat_capacity(P, T)
    Cv = eos.isochoric_heat_capacity(P, T)
    diff = Cp - Cv

    status = "✓" if diff > 0 else "✗"
    print(f"  {desc:<25s}: C_P - C_V = {diff:8.2f} J/(kg·K) {status}")
```

### Test: Maxwell relation (∂S/∂P)_T = -(∂V/∂T)_P = -α/ρ

```python
# Check 2: Maxwell relation (∂S/∂P)_T = -(∂V/∂T)_P = -α/ρ
print("\n2. Maxwell relation check: (∂S/∂P)_T = -α/ρ")
print("-" * 50)
print("   Using numerical derivatives with ΔP = 0.1 GPa, comparing to -α/ρ")

dP = 0.1e9  # 0.1 GPa

for P, T, desc in test_conditions:
    # Numerical (∂S/∂P)_T
    S_plus = eos.specific_entropy(P + dP, T)
    S_minus = eos.specific_entropy(P - dP, T)
    dS_dP_num = (S_plus - S_minus) / (2 * dP)

    # Analytical -α/ρ
    alpha = eos.thermal_expansion(P, T)
    rho = eos.density(P, T)
    dS_dP_analytical = -alpha / rho

    rel_err = abs(dS_dP_num - dS_dP_analytical) / abs(dS_dP_analytical) * 100
    status = "✓" if rel_err < 1.0 else "✗"
    print(f"  {desc:<25s}: num = {dS_dP_num:.4e}, ana = {dS_dP_analytical:.4e}, "
          f"err = {rel_err:.2f}% {status}")
```

### Test: Entropy increases with temperature at constant pressure

```python
# Check 3: dS/dT > 0
print("\n3. Entropy increases with temperature: (∂S/∂T)_P > 0")
print("-" * 50)

P_test_s = [20, 100, 500, 1000]  # GPa
T_test_s = np.linspace(2500, 14000, 50)

all_increasing = True
for P_gpa in P_test_s:
    S_vals = []
    for T in T_test_s:
        try:
            S_vals.append(eos.specific_entropy(P_gpa*GPa, T))
        except RuntimeError:
            S_vals.append(np.nan)
    S_arr = np.array(S_vals)
    dS = np.diff(S_arr[np.isfinite(S_arr)])
    if np.any(dS <= 0):
        all_increasing = False
        print(f"  P = {P_gpa} GPa: ✗ non-monotonic!")
    else:
        print(f"  P = {P_gpa} GPa: ✓ monotonically increasing")

print(f"\n{'All entropy checks passed!' if all_increasing else 'Some entropy checks FAILED.'}") 
```

```python
# Check 4: dρ/dP > 0
print("\n4. Density increases with pressure: (∂ρ/∂P)_T > 0")
print("-" * 50)

T_vals_chk = [2200, 3000, 5000, 8000, 14000]
P_range_chk = np.linspace(5*GPa, 1200*GPa, 100)

all_increasing = True
for T in T_vals_chk:
    rho_vals = []
    for P in P_range_chk:
        try:
            rho_vals.append(eos.density(P, T))
        except RuntimeError:
            rho_vals.append(np.nan)
    rho_arr = np.array(rho_vals)
    drho = np.diff(rho_arr[np.isfinite(rho_arr)])
    if np.any(drho <= 0):
        all_increasing = False
        print(f"  T = {T} K: ✗ non-monotonic!")
    else:
        print(f"  T = {T} K: ✓ monotonically increasing")

print(f"\n{'All density checks passed!' if all_increasing else 'Some density checks FAILED.'}")
```

### Test: Positivity of α and K_T

```python
# Check 5: Positivity of α and K_T
print("\n5. Positivity check: α > 0 and K_T > 0")
print("-" * 50)

P_test = np.linspace(5*GPa, 1000*GPa, 20)
T_test = [3000, 5000, 8000, 14000]

all_positive = True
for T in T_test:
    for P in P_test:
        try:
            alpha = eos.thermal_expansion(P, T)
            # Compute K_T from C_P - C_V = α²T/(ρ) * K_T
            # → K_T = (C_P - C_V) * ρ / (α² T)
            rho = eos.density(P, T)
            Cp = eos.isobaric_heat_capacity(P, T)
            Cv = eos.isochoric_heat_capacity(P, T)
            K_T = (Cp - Cv) * rho / (alpha**2 * T)

            if alpha <= 0:
                print(f"  WARNING: α ≤ 0 at P={P/GPa:.0f} GPa, T={T} K: α = {alpha:.3e}")
                all_positive = False
            if K_T <= 0:
                print(f"  WARNING: K_T ≤ 0 at P={P/GPa:.0f} GPa, T={T} K: K_T = {K_T/GPa:.1f} GPa")
                all_positive = False
        except RuntimeError:
            pass

if all_positive:
    print("  All α > 0 and K_T > 0 across the tested P-T grid ✓")
else:
    print("  Some violations detected ✗")
```

### Test: Grüneisen parameter consistency γ = (C_P − C_V) / (α T C_V)

```python
# Check 6: Grüneisen parameter consistency
# RTpress provides γ internally; we compare with the thermodynamic identity
# γ = (C_P - C_V) / (α T C_V)
print("\n6. Grüneisen parameter: thermodynamic consistency")
print("-" * 50)

# Also verify the adiabatic gradient identity: ∇_ad = α P / (ρ C_P)
# and its relation to γ via ∇_ad = γ T / K_S
# where K_S = C_P/C_V * K_T

print(f"\n  {'Condition':<25s} {'γ_thermo':>10s} {'∇_ad check':>12s} {'Status':>8s}")
print("  " + "-" * 58)

for P, T, desc in test_conditions:
    alpha = eos.thermal_expansion(P, T)
    rho = eos.density(P, T)
    Cp = eos.isobaric_heat_capacity(P, T)
    Cv = eos.isochoric_heat_capacity(P, T)

    gamma_thermo = (Cp - Cv) / (alpha * T * Cv)

    # ∇_ad from class
    nabla_ad = eos.adiabatic_gradient(P, T)
    # ∇_ad from identity: α P / (ρ C_P)
    nabla_check = alpha * P / (rho * Cp)
    nabla_err = abs(nabla_ad - nabla_check) / abs(nabla_ad) * 100

    status = "✓" if nabla_err < 0.1 else "✗"
    print(f"  {desc:<25s} {gamma_thermo:>10.4f} {nabla_err:>10.2f}%  {status:>6s}")
```

## Reference Point Determination <a name="ref-point-mgsio3"></a>

Determine $U_0$ and $S_0$ shifts for all phases so that
$U(P_t, T_t) = 0$ and $S(P_t, T_t) = 0$ at the pyroxene triple
point $(P_t, T_t)$ = (6.5 GPa, 1100 K).

The solid phases that coexist at the triple point (en, lpcen, hpcen)
are evaluated directly. Bridgmanite is matched to HP-CEn at the
12 GPa transition, post-perovskite to bridgmanite at the brg–ppv
boundary, and liquid to the solid phase at the melting curve.

```python
from scipy.optimize import brentq

# Reference point: pyroxene triple point
P0 = 6.5e9   # Pa
T0 = 1100.0  # K

print("=" * 70)
print("Reference Point Shifts for U₀ and S₀")
print("=" * 70)
print(f"Reference point: P₀ = {P0/1e9:.1f} GPa, T₀ = {T0:.1f} K")
print(f"(En–LP-CEn–HP-CEn pyroxene triple point)")
print("=" * 70)

# Dictionary to store results
results = {}

# -------------------------------------------------------------------------
# Pyroxene phases at the triple point (en, lpcen, hpcen)
# -------------------------------------------------------------------------
pyroxene_eos = [
    ("Sokolova22 (en)",    _eos_instances['en']),
    ("Sokolova22 (lpcen)", _eos_instances['lpcen']),
    ("Sokolova22 (hpcen)", _eos_instances['hpcen']),
]

print("\n--- Pyroxene Phases at Triple Point ---\n")
print(f"{'EoS Class':<25} {'U [MJ/kg]':>12} {'S [kJ/(kg·K)]':>15} "
      f"{'U₀ shift [MJ/kg]':>18} {'S₀ shift [kJ/(kg·K)]':>22}")
print("-" * 95)

for name, eos in pyroxene_eos:
    U = eos.specific_internal_energy(P0, T0)
    S = eos.specific_entropy(P0, T0)

    U0_shift = -U
    S0_shift = -S

    results[name] = {
        'U': U, 'S': S,
        'U0_shift': U0_shift, 'S0_shift': S0_shift,
        'P_ref': P0, 'T_ref': T0,
    }

    print(f"{name:<25} {U/1e6:>12.4f} {S/1e3:>15.4f} "
          f"{U0_shift/1e6:>18.4f} {S0_shift/1e3:>22.4f}")

# -------------------------------------------------------------------------
# Bridgmanite: matched to HP-CEn at 12 GPa
# -------------------------------------------------------------------------
print("\n--- Bridgmanite (matched to HP-CEn at 12 GPa) ---\n")

P_brg_ref = 12.0e9  # Pa — HP-CEn/Brg transition
# Use the γ-ε–like logic: evaluate both at the transition pressure
# at a temperature on the phase boundary (same T as the EoS switch).
# Since this is a constant-P boundary, pick T = 1500 K (well within
# both EoS validity ranges).
T_brg_ref = 1500.0  # K

eos_hpcen = _eos_instances['hpcen']
eos_brg   = _eos_instances['brg']

U_hpcen = eos_hpcen.specific_internal_energy(P_brg_ref, T_brg_ref)
S_hpcen = eos_hpcen.specific_entropy(P_brg_ref, T_brg_ref)

U_brg = eos_brg.specific_internal_energy(P_brg_ref, T_brg_ref)
S_brg = eos_brg.specific_entropy(P_brg_ref, T_brg_ref)

# Shift for brg: match shifted HP-CEn at the transition
U0_hpcen_shift = results['Sokolova22 (hpcen)']['U0_shift']
S0_hpcen_shift = results['Sokolova22 (hpcen)']['S0_shift']

U0_brg_shift = (U_hpcen - U_brg) + U0_hpcen_shift
S0_brg_shift = (S_hpcen - S_brg) + S0_hpcen_shift

results['Wolf15 (brg)'] = {
    'U': U_brg, 'S': S_brg,
    'U0_shift': U0_brg_shift, 'S0_shift': S0_brg_shift,
    'P_ref': P_brg_ref, 'T_ref': T_brg_ref,
    'note': 'Continuity with HP-CEn at 12 GPa',
}

print(f"Evaluation point: P = {P_brg_ref/1e9:.0f} GPa, T = {T_brg_ref:.0f} K")
print(f"  HP-CEn:  U = {U_hpcen/1e6:.4f} MJ/kg,  S = {S_hpcen/1e3:.4f} kJ/(kg·K)")
print(f"  Brg:     U = {U_brg/1e6:.4f} MJ/kg,  S = {S_brg/1e3:.4f} kJ/(kg·K)")
print(f"  Jump ΔU = {(U_brg - U_hpcen)/1e6:.4f} MJ/kg, "
      f"ΔS = {(S_brg - S_hpcen)/1e3:.4f} kJ/(kg·K)")
print(f"\n  Required shifts for Brg (to match shifted HP-CEn):")
print(f"  U₀ = {U0_brg_shift/1e6:.4f} MJ/kg,  S₀ = {S0_brg_shift/1e3:.4f} kJ/(kg·K)")

# -------------------------------------------------------------------------
# Post-perovskite: matched to Brg at brg-ppv boundary
# -------------------------------------------------------------------------
print("\n--- Post-perovskite (matched to Brg at brg-ppv boundary) ---\n")

T_ppv_ref = 3000.0  # K
P_ppv_ref = P_brg_ppv(T_ppv_ref)

U_brg_at_ppv = eos_brg.specific_internal_energy(P_ppv_ref, T_ppv_ref)
S_brg_at_ppv = eos_brg.specific_entropy(P_ppv_ref, T_ppv_ref)

eos_ppv = _eos_instances['ppv']
U_ppv = eos_ppv.specific_internal_energy(P_ppv_ref, T_ppv_ref)
S_ppv = eos_ppv.specific_entropy(P_ppv_ref, T_ppv_ref)

U0_ppv_shift = (U_brg_at_ppv - U_ppv) + U0_brg_shift
S0_ppv_shift = (S_brg_at_ppv - S_ppv) + S0_brg_shift

results['Sakai16 (ppv)'] = {
    'U': U_ppv, 'S': S_ppv,
    'U0_shift': U0_ppv_shift, 'S0_shift': S0_ppv_shift,
    'P_ref': P_ppv_ref, 'T_ref': T_ppv_ref,
    'note': 'Continuity with Brg at brg-ppv boundary',
}

print(f"Evaluation point: P = {P_ppv_ref/1e9:.1f} GPa, T = {T_ppv_ref:.0f} K")
print(f"  Brg: U = {U_brg_at_ppv/1e6:.4f} MJ/kg,  S = {S_brg_at_ppv/1e3:.4f} kJ/(kg·K)")
print(f"  PPv: U = {U_ppv/1e6:.4f} MJ/kg,  S = {S_ppv/1e3:.4f} kJ/(kg·K)")
print(f"  Jump ΔU = {(U_ppv - U_brg_at_ppv)/1e6:.4f} MJ/kg, "
      f"ΔS = {(S_ppv - S_brg_at_ppv)/1e3:.4f} kJ/(kg·K)")
print(f"\n  Required shifts for PPv (to match shifted Brg):")
print(f"  U₀ = {U0_ppv_shift/1e6:.4f} MJ/kg,  S₀ = {S0_ppv_shift/1e3:.4f} kJ/(kg·K)")

# -------------------------------------------------------------------------
# Liquid: matched to solid phase at the melting curve
# -------------------------------------------------------------------------
print("\n--- Liquid (matched to solid at melting curve) ---\n")

# Pick T on the melting curve where the liquid EoS converges well
T_liq_ref = 5000.0  # K
P_liq_ref = brentq(lambda P: T_melt_MgSiO3(P) - T_liq_ref, 5e9, 500e9)

print(f"Evaluation point: P = {P_liq_ref/1e9:.1f} GPa, T = {T_liq_ref:.0f} K (on melting curve)")

# Determine the solid phase at this P,T
# Just below the melting curve, this should be brg or ppv
phase_solid = get_mgsio3_phase(P_liq_ref, T_liq_ref - 1.0)
eos_solid = _eos_instances[phase_solid]
print(f"  Solid reference phase: {phase_solid} ({type(eos_solid).__name__})")

# Get the solid shift
solid_key = {
    'hpcen': 'Sokolova22 (hpcen)',
    'brg': 'Wolf15 (brg)',
    'ppv': 'Sakai16 (ppv)',
}[phase_solid]
U0_solid_shift = results[solid_key]['U0_shift']
S0_solid_shift = results[solid_key]['S0_shift']

U_solid = eos_solid.specific_internal_energy(P_liq_ref, T_liq_ref)
S_solid = eos_solid.specific_entropy(P_liq_ref, T_liq_ref)

eos_liq = _eos_instances['liquid']
U_liq = eos_liq.specific_internal_energy(P_liq_ref, T_liq_ref)
S_liq = eos_liq.specific_entropy(P_liq_ref, T_liq_ref)

# Match liquid to shifted solid (ignoring latent heat for now)
U0_liq_shift = (U_solid - U_liq) + U0_solid_shift
S0_liq_shift = (S_solid - S_liq) + S0_solid_shift

results['Wolf18 (liquid)'] = {
    'U': U_liq, 'S': S_liq,
    'U0_shift': U0_liq_shift, 'S0_shift': S0_liq_shift,
    'P_ref': P_liq_ref, 'T_ref': T_liq_ref,
    'note': f'Continuity with {phase_solid} at melting curve (ignoring latent heat)',
}

print(f"  Solid:  U = {U_solid/1e6:.4f} MJ/kg,  S = {S_solid/1e3:.4f} kJ/(kg·K)")
print(f"  Liquid: U = {U_liq/1e6:.4f} MJ/kg,  S = {S_liq/1e3:.4f} kJ/(kg·K)")
print(f"  Raw jump: ΔU = {(U_liq - U_solid)/1e6:.4f} MJ/kg, "
      f"ΔS = {(S_liq - S_solid)/1e3:.4f} kJ/(kg·K)")
print(f"\n  Required shifts for liquid (for continuity, ignoring latent heat):")
print(f"  U₀ = {U0_liq_shift/1e6:.4f} MJ/kg,  S₀ = {S0_liq_shift/1e3:.4f} kJ/(kg·K)")

# -------------------------------------------------------------------------
# Summary: molar values for implementation
# -------------------------------------------------------------------------
print("\n" + "=" * 70)
print("Summary: Molar Shifts for Implementation (J/mol and J/(mol·K))")
print("=" * 70)
print(f"\n{'EoS Class':<25} {'U₀ [kJ/mol]':>14} {'S₀ [J/(mol·K)]':>16}")
print("-" * 58)

for name, data in results.items():
    # Convert from specific (J/kg) to molar (J/mol)
    U0_molar = data['U0_shift'] * _M_MgSiO3
    S0_molar = data['S0_shift'] * _M_MgSiO3
    print(f"{name:<25} {U0_molar/1e3:>14.3f} {S0_molar:>16.3f}")
```

```python
# Generate thermodynamic properties on P-T grid
print("Generating thermodynamic properties on P-T grid...")
print("=" * 80)

# Create grid: log-spaced in P, linear in T
P_grid = np.geomspace(1e-4 * GPa, 1e5 * GPa, 150)
T_grid = np.geomspace(300, 1e5, 150)
TT, PP = np.meshgrid(T_grid, P_grid)

# Properties to compute
properties = [
    ('density', 'kg/m³'),
    ('specific_internal_energy', 'J/kg'),
    ('specific_entropy', 'J/(kg·K)'),
    ('isobaric_heat_capacity', 'J/(kg·K)'),
    ('isochoric_heat_capacity', 'J/(kg·K)'),
    ('thermal_expansion', 'K⁻¹'),
    ('adiabatic_gradient', ''),
]

# Initialize storage
prop_grids = {name: np.full_like(PP, np.nan) for name, _ in properties}
phase_grid_str = np.empty(PP.shape, dtype=object)
n_failures = 0

# Compute properties at each grid point using pre-initialised EoS
for i in range(PP.shape[0]):
    for j in range(PP.shape[1]):
        P, T = PP[i, j], TT[i, j]
        eos, phase = get_eos_for_PT(P, T)
        phase_grid_str[i, j] = phase

        for prop_name, _ in properties:
            try:
                method = getattr(eos, prop_name)
                prop_grids[prop_name][i, j] = method(P, T)
            except RuntimeError:
                n_failures += 1

print(f"✓ Grid computed: {PP.shape[0]} × {PP.shape[1]} = {PP.size:,} points")
print(f"  Non-convergences: {n_failures} ({100*n_failures/PP.size/len(properties):.2f}%)")
```

```python
# Helper function to plot property map with phase boundaries
def plot_property_map_mgsio3(prop_name, units, prop_grid, vmin=None, vmax=None, log_scale=True):
    fig, ax = plt.subplots(figsize=(10, 7), dpi=300)

    # Plot property
    if log_scale:
        data = np.log10(np.abs(prop_grid))
        label = f'log₁₀ |{prop_name}|' + (f' [{units}]' if units else '')
    else:
        data = prop_grid
        label = prop_name + (f' [{units}]' if units else '')

    im = ax.pcolormesh(TT, PP / GPa, data, vmin=vmin, vmax=vmax,
                       shading='auto', cmap='inferno')
    cbar = plt.colorbar(im, ax=ax, label=label)

    # Overlay phase boundaries (black lines)

    # --- Pyroxene boundaries (below triple point, T ≤ 1100 K) ---
    T_low = np.linspace(300, 1100, 200)
    
    # lpcen-en boundary
    P_le = np.array([P_lpcen_en(T) for T in T_low])
    mask_le = (T_low > 750.) # positive-slope branch only
    T_melt_le = np.array([T_melt_MgSiO3(P) for P in P_le])
    mask_le &= (T_low < T_melt_le)
    ax.plot(T_low[mask_le], P_le[mask_le] / GPa, 'w-', linewidth=1.5, alpha=.6)
    
    # lpcen-hpcen boundary
    P_lh = np.array([P_lpcen_hpcen(T) for T in T_low])
    ax.plot(T_low, P_lh / GPa, 'w-', linewidth=1.5, alpha=.6)
    
    # --- Pyroxene boundaries (above triple point, T > 1100 K) ---
    # en-hpcen: extends until it crosses the melting curve
    T_high = np.linspace(1100, 4000, 300)
    P_eh = np.array([P_en_hpcen(T) for T in T_high])
    T_melt_arr = np.array([T_melt_MgSiO3(P) for P in P_eh])
    mask_eh = T_high < T_melt_arr  # stop at melting curve
    ax.plot(T_high[mask_eh], P_eh[mask_eh] / GPa, 'w-', linewidth=1.5, alpha=.6)
    
    # --- hpcen-brg boundary (P = 12 GPa, up to melting) ---
    T_melt_12 = T_melt_MgSiO3(12 * GPa)
    T_hb = np.linspace(300, T_melt_12, 200)
    ax.plot(T_hb, np.full_like(T_hb, 12.0), 'w-', linewidth=1.5, alpha=.6)
    
    # --- brg-ppv boundary: extends until melting curve ---
    T_ppv = np.linspace(300, 1e5, 300)
    P_bp = np.array([P_brg_ppv(T) for T in T_ppv])
    T_melt_bp = np.array([T_melt_MgSiO3(P) for P in P_bp])
    mask_bp = (T_ppv < T_melt_bp) & (P_bp > 0)
    ax.plot(T_ppv[mask_bp], P_bp[mask_bp] / GPa, 'w-', linewidth=1.5, alpha=.6)
    
    # --- Melting curve ---
    P_melt = np.logspace(np.log10(1e-4 * GPa), np.log10(1e5 * GPa), 500)
    T_melt = np.array([T_melt_MgSiO3(P) for P in P_melt])
    ax.plot(T_melt, P_melt / GPa, 'w-', linewidth=1.5, alpha=.6)

    ax.loglog()
    ax.set_xlabel('Temperature [K]', fontsize=12)
    ax.set_ylabel('Pressure [GPa]', fontsize=12)
    ax.set_xlim(300, 1e5)
    ax.set_ylim(1e-4, 1e5)

    plt.tight_layout()
    plt.show()
```

```python
# Compute thermodynamic consistency parameter
# Delta = 1 - rho^2 (dS/dP)_T / (drho/dT)_P
# For a thermodynamically consistent EoS, Delta = 0 (Maxwell relation)

Delta = np.full_like(PP, np.nan)
phase_map_delta = np.empty(PP.shape, dtype=object)

# Finite difference steps (relative)
dP_rel = 1e-4
dT_rel = 1e-4

for i in range(PP.shape[0]):
    for j in range(PP.shape[1]):
        P, T = PP[i, j], TT[i, j]

        try:
            eos, phase = get_eos_for_PT(P, T)
            phase_map_delta[i, j] = phase

            # Central differences for derivatives
            dP = P * dP_rel
            dT = T * dT_rel

            # (dS/dP)_T
            S_plus = eos.specific_entropy(P + dP, T)
            S_minus = eos.specific_entropy(P - dP, T)
            dS_dP = (S_plus - S_minus) / (2 * dP)

            # (drho/dT)_P
            rho_plus = eos.density(P, T + dT)
            rho_minus = eos.density(P, T - dT)
            drho_dT = (rho_plus - rho_minus) / (2 * dT)

            # Density at current point
            rho = eos.density(P, T)

            # Consistency parameter
            if abs(drho_dT) > 1e-20:
                Delta[i, j] = 1 - rho**2 * dS_dP / drho_dT

        except Exception:
            phase_map_delta[i, j] = None
            continue
```

```python
# Statistics on thermodynamic consistency parameter Delta

print("=" * 70)
print("Thermodynamic Consistency: Δ = 1 - ρ² (∂S/∂P)_T / (∂ρ/∂T)_P")
print("Perfect consistency: Δ = 0")
print("=" * 70)

# Overall statistics
valid_mask = ~np.isnan(Delta)
Delta_valid = Delta[valid_mask]

print(f"\n{'Overall Statistics':^70}")
print("-" * 70)
print(f"Total grid points:    {Delta.size:>10}")
print(f"Valid points:         {valid_mask.sum():>10} ({100*valid_mask.sum()/Delta.size:.1f}%)")
print(f"Invalid/failed:       {(~valid_mask).sum():>10}")

print(f"\n{'Statistic':<20} {'Value':>15} {'|Δ|':>15}")
print("-" * 52)
print(f"{'Mean':<20} {np.mean(Delta_valid):>15.2e} {np.mean(np.abs(Delta_valid)):>15.2e}")
print(f"{'Median':<20} {np.median(Delta_valid):>15.2e} {np.median(np.abs(Delta_valid)):>15.2e}")
print(f"{'Std':<20} {np.std(Delta_valid):>15.2e} {np.std(np.abs(Delta_valid)):>15.2e}")
print(f"{'Min':<20} {np.min(Delta_valid):>15.2e}")
print(f"{'Max':<20} {np.max(Delta_valid):>15.2e}")

# Percentiles
print(f"\n{'Percentiles of |Δ|':^70}")
print("-" * 70)
percentiles = [50, 90, 95, 99, 99.9]
abs_Delta = np.abs(Delta_valid)
for p in percentiles:
    val = np.percentile(abs_Delta, p)
    print(f"  {p:>5.1f}%:  |Δ| ≤ {val:.2e}")

# Consistency thresholds
print(f"\n{'Consistency Assessment':^70}")
print("-" * 70)
thresholds = [1e-10, 1e-8, 1e-6, 1e-4, 1e-2, 0.1, 1.0]
for thresh in thresholds:
    count = np.sum(abs_Delta <= thresh)
    pct = 100 * count / len(abs_Delta)
    print(f"  |Δ| ≤ {thresh:<8.0e}:  {count:>8} points ({pct:>6.2f}%)")

# Per-phase statistics
print(f"\n{'Statistics by Phase':^70}")
print("-" * 70)
phases = ['en', 'lpcen', 'hpcen', 'brg', 'ppv', 'liquid']
print(f"{'Phase':<12} {'Count':>8} {'Mean |Δ|':>12} {'Median |Δ|':>12} {'Max |Δ|':>12}")
print("-" * 60)

for phase in phases:
    mask = (phase_map_delta == phase) & valid_mask
    if mask.sum() > 0:
        Delta_phase = Delta[mask]
        abs_Delta_phase = np.abs(Delta_phase)
        print(f"{phase:<12} {mask.sum():>8} {np.mean(abs_Delta_phase):>12.2e} "
              f"{np.median(abs_Delta_phase):>12.2e} {np.max(abs_Delta_phase):>12.2e}")
    else:
        print(f"{phase:<12} {0:>8} {'N/A':>12} {'N/A':>12} {'N/A':>12}")

# Flag problematic regions
print(f"\n{'Problematic Regions (|Δ| > 0.01)':^70}")
print("-" * 70)
problem_mask = np.abs(Delta) > 0.01
if problem_mask.sum() > 0:
    P_prob = PP[problem_mask]
    T_prob = TT[problem_mask]
    Delta_prob = Delta[problem_mask]
    phase_prob = phase_map_delta[problem_mask]

    for phase in phases:
        phase_mask = phase_prob == phase
        if phase_mask.sum() > 0:
            P_range = (P_prob[phase_mask].min()/GPa, P_prob[phase_mask].max()/GPa)
            T_range = (T_prob[phase_mask].min(), T_prob[phase_mask].max())
            print(f"  {phase}: {phase_mask.sum()} points")
            print(f"    P: {P_range[0]:.1f} - {P_range[1]:.1f} GPa")
            print(f"    T: {T_range[0]:.0f} - {T_range[1]:.0f} K")
else:
    print("  None found — all points have |Δ| ≤ 0.01")
```
